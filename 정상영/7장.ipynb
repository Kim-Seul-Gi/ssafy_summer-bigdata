# 7장 시계열(Time Series) 데이터를 다뤄보자

```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import pandas_datareader.data as web
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from fbprophet import Prophet
from datetime import datetime
```

```python
pinkwink_web = pd.read_csv('../data/08. PinkWink Web Traffic.csv',encoding='utf-8', thousands=',',
                          names = ['date','hit'], index_col=0)
pinkwink_web = pinkwink_web[pinkwink_web['hit'].notnull()]
pinkwink_web.head()
```

```python
pinkwink_web['hit'].plot(figsize=(12,4), grid=True);
```

```python
time = np.arange(0,len(pinkwink_web))
traffic = pinkwink_web['hit'].values

fx = np.linspace(0, time[-1], 1000)
```

```python
def error(f, x, y):
    return np.sqrt(np.mean((f(x)-y)**2))
```

```python
fp1 = np.polyfit(time, traffic, 1)
f1 = np.poly1d(fp1)

f2p = np.polyfit(time, traffic, 2)
f2 = np.poly1d(f2p)

f3p = np.polyfit(time, traffic, 3)
f3 = np.poly1d(f3p)

f15p = np.polyfit(time, traffic, 15)
f15 = np.poly1d(f15p)

print(error(f1, time, traffic))
print(error(f2, time, traffic))
print(error(f3, time, traffic))
print(error(f15, time, traffic))
```

```python
plt.figure(figsize=(10,6))
plt.scatter(time, traffic, s=10)

plt.plot(fx,f1(fx), lw=4, label='f1')
plt.plot(fx,f2(fx), lw=4, label='f2')
plt.plot(fx,f3(fx), lw=4, label='f3')
plt.plot(fx,f15(fx), lw=4, label='f15')

plt.grid(True, linestyle='-', color='0.75')

plt.legend(loc=2)
plt.show()
```

### 7-2 Prophet 모듈을 이용한 forecast 예측

Prophet 모듈은 사용법이 아주 간단하다. pinkwink_web 변수에서 날짜(index)와 방문수(hit)만 따로 저장한다.

```python
df = pd.DataFrame({'ds':pinkwink_web.index, 'y':pinkwink_web['hit']})
df.reset_index(inplace=True)
df['ds'] = pd.to_datetime(df['ds'], format="%y. %m. %d.")
del df['date']

m = Prophet(yearly_seasonality=True)
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=60)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
m.plot(forecast)
```

```python
m.plot_components(forecast)
```

해당 블로그는 월, 화, 수요일 방문자가 많고 주말은 아주 낮다. 그리고 마지막으로 재미있는 것이 블로그는 3월부터 상승해서 7월쯤 내려가기 시작한다. 5월에 한 번 숨고르기를 하고 다시 9얼경 상승해서 12월경 내려온다. 아무래도 공대에서 좋아하는 포스팅이 많아서인가보다. Prophet을 사용해서 간략하게 흥미로운 결과를 얻었다. 앞으로의 예측부터 주별, 연간 데이터의 현황도 알 수 있게 되었다. 비록 빈약하긴 하지만 이러한 과정을 Seasonal 시계열 분석이라고한다.

```python
df = pd.read_csv('../data/08. example_wp_peyton_manning.csv')
df['y'] = np.log(df['y'])
m = Prophet(daily_seasonality=True)
m.fit(df)
future = m.make_future_dataframe(periods=366)
```

```python
df.y.plot(figsize=(12,6), grid=True);
```

```python

playoffs = pd.DataFrame({
  'holiday': 'playoff',
  'ds': pd.to_datetime(['2008-01-13', '2009-01-03', '2010-01-16',
                        '2010-01-24', '2010-02-07', '2011-01-08',
                        '2013-01-12', '2014-01-12', '2014-01-19',
                        '2014-02-02', '2015-01-11', '2016-01-17',
                        '2016-01-24', '2016-02-07']),
  'lower_window': 0,
  'upper_window': 1,
})
superbowls = pd.DataFrame({
  'holiday': 'superbowl',
  'ds': pd.to_datetime(['2010-02-07', '2014-02-02', '2016-02-07']),
  'lower_window': 0,
  'upper_window': 1,
})
holidays = pd.concat((playoffs, superbowls))
```

```python
m = Prophet(holidays=holidays, daily_seasonality=True)
forecast = m.fit(df).predict(future)
```

```python
forecast[(forecast['playoff'] + forecast['superbowl']).abs() > 0][
        ['ds', 'playoff', 'superbowl']][-10:]
```

```python
m.plot(forecast)
```

```python
m.plot_components(forecast);
```
